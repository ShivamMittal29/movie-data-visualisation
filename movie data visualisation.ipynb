```python
titles=list(data.get('movie_title'))
print(titles[:5])
```

```python
freq_title={}
for title in titles:
    length=len(title)
    
    if freq_title.get(length) is None:
        freq_title[length]=1
    else:
        freq_title[length]+=1
```

```python
freq_title
```

```python
X=np.array(list(freq_title.keys()))
Y=np.array(list(freq_title.values()))
```

```python
plt.scatter(X,Y)
plt.xlabel("length of the movie title")
plt.ylabel("the number of movies having names of the given length")
plt.title("Movie data visualisation")
plt.show()
```
